```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'  # Windows에서는 'Malgun Gothic', macOS에서는 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# 그래프 생성
G = nx.DiGraph()

# 노드 추가
nodes = {
    "i1": {"pos": (0, 2), "label": "$i_1$", "color": "white"},
    "i2": {"pos": (0, 1), "label": "$i_2$", "color": "white"},
    "h1": {"pos": (2, 2), "label": "$b_{11}$", "color": "#FFD1D1"},  # 레이어1 노드
    "h2": {"pos": (2, 1), "label": "$b_{12}$", "color": "#FFD1D1"},  # 레이어1 노드
    "o": {"pos": (4, 1.5), "label": "$b_{21}$", "color": "#FFD1D1"}  # 출력 노드
}

edges = [
    ("i1", "h1", "$w_{11}$"),
    ("i1", "h2", "$w_{12}$"),
    ("i2", "h1", "$w_{21}$"),
    ("i2", "h2", "$w_{22}$"),
    ("h1", "o", "$w_1$"),
    ("h2", "o", "$w_2$")
]

# 그래프 그리기
plt.figure(figsize=(10, 6))
ax = plt.gca()
ax.set_facecolor("#FFECEC")  # 배경 색상 설정

# 노드 그리기
for node, attr in nodes.items():
    x, y = attr["pos"]
    circle = plt.Circle((x, y), radius=0.3, color=attr["color"], ec="black", zorder=3)
    ax.add_artist(circle)
    plt.text(x, y, attr["label"], fontsize=10, ha="center", va="center")

# 엣지 그리기
for edge in edges:
    start_pos = nodes[edge[0]]["pos"]
    end_pos = nodes[edge[1]]["pos"]
    ax.annotate("", xy=end_pos, xytext=start_pos,
                arrowprops=dict(arrowstyle="->", color="black"))
    mid_x = (start_pos[0] + end_pos[0]) / 2
    mid_y = (start_pos[1] + end_pos[1]) / 2
    plt.text(mid_x, mid_y, edge[2], fontsize=8, color="black")

# 레이어 텍스트 추가
plt.text(-0.5, 1.5, "입력", fontsize=12, color="black")
plt.text(2.5, 3.0, "레이어 1", fontsize=12, color="black")
plt.text(4.5, 2.5, "출력", fontsize=12, color="black")

# 축 숨기기 및 설정
ax.set_xlim(-1, 5)
ax.set_ylim(0, 3)
ax.axis("off")
plt.title("신경망의 활성함수", fontsize=14)
plt.show()
```
